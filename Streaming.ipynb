```python
import pandas as pd
import os
import kagglehub
```

```python
path = r"C:\Users\nickh\.cache\kagglehub\datasets\mlg-ulb\creditcardfraud\versions\3"

df = pd.read_csv(os.path.join(path, "creditcard.csv"))

print(df.head())
print("\nColumns:", df.columns.tolist())
print("\nShape:", df.shape)
```

Sort by time

```python
df_sorted = df.sort_values("Time").reset_index(drop=True)
events = []

for i, row in df_sorted.head(10).iterrows():
    event = {
        "event_id": i,
        "event_time": float(row["Time"]),
        "amount": float(row["Amount"]),
        "features": [float(row[f"V{j}"]) for j in range(1, 29)],
        "label": int(row["Class"])
    }
    events.append(event)

for e in events:
    print(e)
```

```python
import time
from datetime import datetime

for event in events[:5]:
    print(f"[{datetime.now().strftime('%H:%M:%S')}] sending event_id={event['event_id']} "
          f"event_time={event['event_time']} amount={event['amount']}")
    time.sleep(1)
```
